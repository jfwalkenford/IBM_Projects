### Is there a significant difference in median value of houses bounded by the Charles river or not?

Hypothesis: There will by a difference between the homes bound by the Charles river and those not bound by the Charles River.

Null(H<sub>0</sub>) Hypothesis:There will be no significant difference between the Value of the Owner Occupied Homes and residing near the Charles River.

Alternative(H<sub>1</sub>) Hypothesis:There will be a significant difference between the Value of the Owner Occupied Homes and residing near the Charles River.

```python
scipy.stats.levene(boston_df[boston_df['CHAS'] == 0]['MEDV'],
                   boston_df[boston_df['CHAS'] == 1]['MEDV'], center='mean')
```

Since the pvalue is less than 0.05 we can assume unequal variance.

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 0]['MEDV'],
                   boston_df[boston_df['CHAS'] == 1]['MEDV'])
```

Since the p-value is less than alpha value 0.05, we reject the null hypothesis as there is enough proof that there is a statistical difference in Charles River owners and the Median Value of Owner Occupied Homes

```python
scipy.stats.levene(
    boston_df[boston_df['antique'] == 'youngest']['MEDV'],
    boston_df[boston_df['antique'] == 'balanced']['MEDV'],
    boston_df[boston_df['antique'] == 'oldest']['MEDV'], center='mean'
)
```

Since the pvalue is greater than 0.05 we can assume equal variances among the three groups.

```python
thirty_five_younger = boston_df[boston_df['antique'] == 'youngest']['MEDV']
thirt_five_seventy = boston_df[boston_df['antique'] == 'balanced']['MEDV']
seventy_older = boston_df[boston_df['antique'] == 'oldest']['MEDV']
```

```python
f_statistic, p_value = scipy.stats.f_oneway(thirty_five_younger, thirt_five_seventy, seventy_older)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))
```

Since the P-Value is lower than 0.05 the Null Hypothesis is rejected and it is concluded that there is a difference between median values of houses and at least two proportions of owner occupied units built prior to 1940.

### Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town?

Hypothesis: There will be no relationship between NO<sub>2</sub> and the proportion of the non-retail business acres.

Null(H<sub>0</sub>) Hypothesis:There will be no significant difference between the Concentration of Nirtic Oxide and the Non-retail business acres per town.

Alternative(H<sub>1</sub>) Hypothesis:There will be a relationship between the Concentration of Nirtic Oxide and the Non-retail business acres per town.

```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

Since the P-Value is below 0.05 the Null Hypothesis has been rejected and there is suffiecent evidence to accept that there is a significant relationship between the Concentration of Nirtric Oxide and the number of Non-retail business acres per town.

### What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes?

Hypothesis: The distance to the Boston employment centres will have an impact on the owner occupied homes.

Null(H0) Hypothesis:There will be no significant difference between the distance of the 5 employment centres on the median value of owner occupied homes

Alternative(H1) Hypothesis:There will be significant difference between the distance of the 5 employment centres on the median value of owner occupied homes

```python
X = boston_df['DIS']
y = boston_df['MEDV']
## add an intercept (beta_0)
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

model.summary()
```

With the P-value is lower than 0.05 the null hypothesis is rejected and the correlation between distance to the Boston Employment Centres and the median value is correlated.
